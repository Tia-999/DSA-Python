```python
from collections import deque
```

```python
class Queue:
    def __init__(self):
        self.buffer = deque()
    def enqueue(self,val):
        self.buffer.appendleft(val)
    def dequeue(self):
        return self.buffer.pop()
    def is_empty(self):
        return len(self.buffer) == 0
    def size(self):
       return len(self.buffer)
    def front(self):
        return self.buffer[-1]
    
```

Serve Order: This thread will server the order. All you need to do is pop the order out of the queue and print it. This thread serves an order every 2 seconds. Also start this thread 1 second after place order thread is started.

```python
orders = ['pizza','samosa','pasta','biryani','burger']
served = []
food = Queue()
def place_order(arr):
    for order in arr:
        print(f"Placing order: {order}")
        food.enqueue(order)
        time.sleep(0.5)
        
def serve_order(arr):
    while True:
        if not food.is_empty():
            order  = food.dequeue()
            served.append(order)
            print(f"Served: {order}")
        time.sleep(2)
        
t = time.time()
t1 = threading.Thread(target = place_order, args =(orders,))
t2 = threading.Thread(target = serve_order, args =(orders,))
t1.start()
time.sleep(1)
t2.start()
t1.join()
t2.join()
```

Write a program to print binary numbers from 1 to 10 using Queue. Use Queue class implemented in main tutorial. Binary sequence should look like,
    1
    10
    11
    100
    101
    110
    111
    1000
    1001
    1010

```python
def produce_binary_numbers(n):
    n_q = Queue()
    n_q.enqueue("1")
    for i in range(n):
        front = n_q.front()
        print(f"  {front}")
        n_q.enqueue(front+"0")
        n_q.enqueue(front+"1")
        n_q.dequeue()

produce_binary_numbers(20)
```
