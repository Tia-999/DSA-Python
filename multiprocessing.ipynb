```python
import multiprocessing
```

```python
def calc_square(numbers, result,v):
    v.value = 7.8
    for idx,n in enumerate(numbers):
        result[idx] = n*n

numbers =[7,8,9]
result = multiprocessing.Array('i',3) #integer and 3
v = multiprocessing.Value('d',0.0)
p = multiprocessing.Process(target = calc_square, args =(numbers,result,v))
p.start()
p.join()
print(v.value)
print(result[:])
```

Between 2 Processes -Queue

```python
def calc_square(numbers, result):
    for n in numbers:
        result.put(n*n)

numbers =[7,8,9]
result = multiprocessing.Queue()
p = multiprocessing.Process(target = calc_square, args =(numbers,result))
p.start()
p.join()

while result.empty() is False:
    print(result.get())
```
