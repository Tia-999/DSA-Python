# LINKED LIST


```python
class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next
class LinkedList:
    def __init__(self):
        self.head = None
    def insert_at_begining(self,data):
        node = Node(data,self.head)
        self.head = node
        
    def printll(self):
        if self.head == None:
            print("empty ll")
        ptr = self.head
        llstr = ""
        while ptr:
            llstr += str(ptr.data) + "---->"
            ptr = ptr.next
        print(llstr)
        
    def insert_at_end(self,data):
        
        ptr = self.head
        if ptr is None:
            ptr = Node(data,None)
            return
        while ptr.next:
            ptr = ptr.next
        ptr.next = Node(data,None)
    def insert_list(self,data_list):
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
    def gl(self):
        c = 0
        ptr = self.head
        while ptr:
            c += 1
            ptr = ptr.next
        return c
        
    def remove_at(self, index):
        if index < 0 and index > self.gl():
            raise Exception("Invalid Index")
        if index == 0:
            self.head = self.head.next
        ptr = self.head
        count = 0
        while ptr:
            if count == index-1:
                ptr.next = ptr.next.next
                break
            ptr = ptr.next
            count += 1
            
    def insert_at(self, index,data):
        if index < 0 and index > self.gl():
            raise Exception("Invalid Index")
        if index == 0:
            self.insert_at_begining(data)

        ptr = self.head
        count = 0
        while ptr:
            if count == index-1:
                node = Node(data,ptr.next)
                ptr.next = node
                break
            ptr = ptr.next
            count += 1
            
    def insert_after_value(self, data_after, data_to_insert):
        ptr = self.head
        while ptr:
            if ptr.data == data_after:
                node = Node(data_to_insert,ptr.next)
                ptr.next = node
            ptr = ptr.next
            
    def remove_by_value(self,data):
        ptr = self.head
        while ptr.next:
            if ptr.next.data == data:
                ptr.next = ptr.next.next
                break
            ptr = ptr.next
                
                
        

            
        
```

# Doubly Linked List

```python

class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    def insert_at_begining(self,data):
         if self.head== None:
             node = Node(data,self.head,None)
             self.head = node
         else:
             node = Node(data,self.head,None)
             self.head.prev = node
             self.head = node
    def get_length(self):
        c = 0
        ptr = self.head
        while ptr:
            c += 1
            ptr = ptr.next
        return c
    def print_forward(self):
        ptr = self.head
        llstr = ' '
        while ptr:
            llstr += str(ptr.data) + '------>'
            ptr = ptr.next
        print(llstr)
    def print_backward(self):
        if self.head is None:
            print("Linked list is empty")
            return
             
        last_node = self.gln()
        ptr = last_node
        llstr = ' '
        while ptr:
               llstr += str(ptr.data) + '------>'
               ptr = ptr.prev
        print(llstr)
    
    def gln(self):
        ptr = self.head
        while ptr.next:
            ptr= ptr.next
        return ptr

    def insert_at_end(self,data):
        
        last_node = self.gln()
        if self.head is None:
            self.head = Node(data, None, None)
            return
        else:
            node = Node(data,None,last_node)
            last_node.next = node
            
    def insert_at(self, index, data):
        if index<0 or index>self.get_length():
            raise Exception("Invalid Index")

        if index==0:
            self.insert_at_begining(data)
            return
        c = 0
        ptr = self.head
        while ptr:
            if c == index -1:
               node = Node(data,ptr.next,ptr)
               if ptr.next:
                  ptr.next.prev = node
               ptr.next = node
               break
            ptr = ptr.next
            c +=1
    def remove_at(self, index):
        if index<0 or index>self.get_length():
            raise Exception("Invalid Index")

        if index==0:
            self.head= self.head.next
            self.head.prev = None
            return
            
        c = 0
        ptr = self.head
        while ptr:
            if c == index :
               ptr.prev.next = ptr.next
               if ptr.next:
                   ptr.next.prev = ptr.prev
               break
                
            ptr = ptr.next
            c +=1
            
            
            
        
        
            
    
        
            
        
```

```python
dl = DoublyLinkedList()
dl.insert_at_begining(55)
dl.insert_at_begining(77)
dl.insert_at_begining(89)
dl.insert_at_begining(66)
dl.insert_at_end(699)
dl.insert_at(3,26)
dl.remove_at(2)
dl.print_forward()
dl.print_backward()
```
